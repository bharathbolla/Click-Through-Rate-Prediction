```python
# import required packages
from pathlib import Path
from pprint import pprint
from joblib import dump, load
from collections import OrderedDict
from IPython.display import display, Markdown, Latex, HTML
from pandas.api.types import is_datetime64_any_dtype as is_datetime



import os
import time
import pickle
import datetime
import warnings
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import category_encoders as ce


from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import sklearn.metrics as skm
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn import set_config
from sklearn.utils import estimator_html_repr


import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import plotly.express as px

rs = {'random_state': 42}


from sklearn.compose import make_column_selector as selector
```

```python
def returnCatNumList(df):
    
    object_cols = list(df.select_dtypes(exclude=np.number).columns)
    numeric_cols = list(df.select_dtypes(include=np.number).columns)

    return object_cols, numeric_cols

# cat_cols, num_cols = returnCatNumList(diamonds)


def get_model_metrics(y_test, y_pred):

    # F1-Score
    fscore = skm.f1_score(y_test, y_pred)
    # Precision
    pres = skm.precision_score(y_test, y_pred)
    # Recall
    rcall = skm.recall_score(y_test, y_pred)
    # Accuracy
    accu = skm.accuracy_score(y_test, y_pred)
    # ROC_AUC
    roc_auc = skm.roc_auc_score(y_test, y_pred)

    return fscore, pres, rcall, accu, roc_auc



def categorical_encoding_benchmark(X, y, num_cols):


    # cat enc instances
    encoder_cat = ce.CatBoostEncoder()
    encoder_cnt = ce.CountEncoder()
    encoder_loo = ce.LeaveOneOutEncoder()
    #encoder_ohe = ce.OneHotEncoder()
    encoder_ord = ce.OrdinalEncoder()
    encoder_tar = ce.TargetEncoder()
    encoder_woe = ce.WOEEncoder()

    print("----------------------------->> Category Encoders instances created")

    encoders_list = [
                     ("CatBoostEncoder", encoder_cat),
                     ("CountEncoder", encoder_cnt),
                     ("LeaveOneOutEncoder", encoder_loo), 
                     #("OneHotEncoder", encoder_ohe), 
                     ("OrdinalEncoder", encoder_ord),
                     ("TargetEncoder",encoder_tar), 
                     ("WOEEncoder",encoder_woe)]


    log_reg = LogisticRegression(solver='lbfgs',
                                 max_iter=10000,
                                 **rs)
    dt = DecisionTreeClassifier(**rs)
    rf = RandomForestClassifier(**rs)
    xgb = XGBClassifier(**rs, 
                        verbosity=0, 
                        use_label_encoder=False)


    clfs = [
            ('Logistic Regression', log_reg), 
            ('Random Forest', rf), 
            ('XGBoost', xgb)
            ]

    print("------------------------------------->> ML Models instances created")
    pipelines = {
                 'CatBoostEncoder': [], 
                 'CountEncoder': [],
                 'LeaveOneOutEncoder': [],
                 #'OneHotEncoder': [], 
                 'OrdinalEncoder': [], 
                 'TargetEncoder': [], 
                 'WOEEncoder': []}

    scores_df = pd.DataFrame(columns=['Categorical_Encoder', 'Model', 'F1_Score', 
                                      'Precision', 'Recall', 'Accuracy', 'ROC_AUC',
                                      'Timetaken_to_Mdl(secs)'])
    



    # Train-test Split
    X_train, X_test, y_train, y_test = train_test_split(X, 
                                                        y, 
                                                        test_size = 0.30, 
                                                        random_state = 42,  stratify =y)
    print("Data Split --------------------->> X_train, X_test, y_train, y_test")
    
    from imblearn.over_sampling import RandomOverSampler
    randomsample=  RandomOverSampler()
    X_train,y_train=randomsample.fit_resample(X_train,y_train)

    
    for encoder_name, encoder in encoders_list:

        print(encoder_name)


        numeric_preprocessor = Pipeline(
            steps=[
                   ("imputation_mean", SimpleImputer(missing_values=np.nan, 
                                                     strategy="mean")),
                   ("scaler", StandardScaler()),
            ]
        )

        categorical_preprocessor = Pipeline(
            steps=[
                   ("imputation_constant",
                    SimpleImputer(fill_value="missing", strategy="constant"),
                ),
                (encoder_name, encoder),
            ]
        )


        numerical_columns_selector = selector(dtype_exclude=object)
        categorical_columns_selector = selector(dtype_include=object)

        numerical_columns = numerical_columns_selector(X)
        categorical_columns = categorical_columns_selector(X)

        preprocessor = ColumnTransformer(
            [
                ("categorical", categorical_preprocessor, categorical_columns),
                ("numerical", numeric_preprocessor, numerical_columns),
            ]
        )

        
        for clf_name, clf in clfs:

            clf_start_time = time.time()

            print(clf_name)

            pipeline = Pipeline(
                steps=[
                       ("preprocessor", preprocessor), 
                       ("classifier", clf)
                       ]
            )


            pipeline.fit(X_train, y_train)


            y_pred = pipeline.predict(X_test)

            
            fscore, pres, rcall, accu, roc_auc = get_model_metrics(y_test, y_pred)


            # break
            time_taken_to_mdl = time.time() - clf_start_time
            # time_fmt = time.strftime("%H:%M:%S",time.gmtime(seconds))

            pipelines[encoder_name].append(pipeline)
                
            scores_dict = {
                'Categorical_Encoder': encoder_name,
                'Model' : clf_name, 
                'F1_Score' : fscore,
                'Precision' : pres,
                'Recall' : rcall,
                'Accuracy' : accu,
                'ROC_AUC' : roc_auc,
                'Timetaken_to_Mdl(secs)':time_taken_to_mdl
                }
            
            scores_df = pd.concat([scores_df, pd.DataFrame.from_records([scores_dict])])

            print("--------------------------->> Evaluation of Model completed")

            print(f"ML Pipeline Successfully----------------->> {encoder_name}")
        
        print("--------------------->>>>>>>Successfully completed CE Benchmark")
            
    return pipelines, scores_df



def summary_benchmark_models(scores_df, color_discrete_map, filename):


    fig = px.bar(
        scores_df,
        y="Model",
        x="F1_Score",
        color="Categorical_Encoder",
        color_discrete_map=color_discrete_map,
        custom_data=['Precision', 'Recall',
                    'Accuracy', 'ROC_AUC'],
        orientation='h'
        )

    fig.update_traces(
        texttemplate="%{x:.2f}",
        textposition="inside",
        insidetextanchor="middle",
        hovertemplate="<br>".join(
            ["Model: %{y}",
            "F1 Score: %{x:.2f}",
            "Precision: %{customdata[0]:.2f}",
            "Recall: %{customdata[1]:.2f}",
            "Accuracy: %{customdata[2]:.2f}",
            "ROC_AUC: %{customdata[3]:.2f}"]
                )
        )

    fig.update_layout(
        title = {'text': "Summary Metrics",
                'x':0.5, 
                'xanchor': 'center'},
        title_font_size=20,
        yaxis=dict(
            title='Models',
            titlefont_size=16
            ),
        hoverlabel=dict(
            bgcolor="white",
            font_size=16,
            font_family="Rockwell"
        ),
        xaxis=dict(
            title="F1-Score",
            titlefont_size=16,

        ),
        legend_title="Categorical Encoders",
        height=600,


    )

    # fig.show()
    fig.write_html(f"CE_BenchMark/Figures/Plotly/{filename}.html")


def CE_BenchMark_Pipeline(datasets, targets):

    #---------------------------------------------------------------------------
    # Create required folders
    Path(f"CE_BenchMark/Figures/Plotly/").mkdir(parents=True, 
                                exist_ok=True)
    
    Path(f"CE_BenchMark/Pipelines/").mkdir(parents=True, 
                                           exist_ok=True)

    Path(f"CE_BenchMark/Scores/").mkdir(parents=True, 
                                        exist_ok=True)

    Path(f"CE_BenchMark/Metadata/").mkdir(parents=True, 
                                        exist_ok=True)
    #---------------------------------------------------------------------------

    file_trail = f"{str(now.day)}_{str(now.month)}_{str(now.year)}_{str(now.hour)}_{str(now.minute)}_{str(now.second)}_"


    pipeline_dict = OrderedDict()
    #---------------------------------------------------------------------------

    metadata  = OrderedDict()

    for df_name, df in datasets.items():

        print(df_name)

        

        # try:

        # Drop datetime
        date_cols = [col for col in df.columns if is_datetime(col)]

        if len(date_cols) > 0:
            data = df.drop(columns=date_cols, errors=False)
        else:
            data = df.copy()

        tar = targets[df_name]
        
        cat_cols, num_cols = returnCatNumList(data)

        num_cols.remove(tar)

    #---------------------------------------------------------------------------

        

        X = data.drop(columns=tar)
        y = data[tar]

        
        pipelines, scores_df = categorical_encoding_benchmark(X, y, num_cols)




        print(f"---------->> Categorical Benchmark completed for {df_name}")

        dump(pipelines,f"CE_BenchMark/Pipelines/{df_name}_{file_trail}.joblib")

        # scores_df.to_csv(f"CE_BenchMark/Scores/{file}_{file_trail}.csv", index=False)

        cm = sns.color_palette("Set2", as_cmap=True)
        
        with pd.ExcelWriter(f"CE_BenchMark/Scores/Scores_styled_{file_trail}.xlsx") as writer:  
            # scores_df.to_excel(writer, sheet_name=file)
            scores_df.set_index(
                ['Categorical_Encoder','Model']
                ).style.background_gradient(cmap=cm).to_excel(
                                writer,
                                sheet_name=df_name
                                )


        fig_sv_name = f"{df_name}_{file_trail}"
        colors_ = "#66C5CC,#F6CF71,#F89C74,#DCB0F2,#87C55F,#9EB9F3,#FE88B1,#C9DB74,#8BE0A4,#B497E7,#D3B484,#B3B3B3,#39b185,#9ccb86,#e9e29c,#eeb479,#e88471"
        colors = colors_.split(",")
        col_values = scores_df['Categorical_Encoder'].unique()
        color_discrete_map = dict(zip(col_values, colors[: len(col_values)]))
        summary_benchmark_models(scores_df, color_discrete_map, fig_sv_name)
        print(scores_df)
        cate_prop = len(cat_cols)/X.shape[1]
        num_prop = len(num_cols)/X.shape[1]

        meta = {"Dataset":df_name,
                # "TimeTaken(min)":scores_df["Time_taken"].sum()/60,
                "Categorical_Columns": cat_cols, 
                "Numeric_Columns": num_cols,
                "n_cate_cols": len(cat_cols),
                "n_num_cols": len(num_cols),
                "total_cols": X.shape[1],
                "cate_prop": cate_prop,
                "num_prop":num_prop
                }

        metadata[df_name] = meta
        pipeline_dict[df_name] = pipelines

        print("===="*30)
        print(df_name)
        print("===="*30)
        pprint(meta)



        # except Exception as E:
        #     print("===="*30)
        #     print(f"{df_name}------>Failed")
        #     print()
        #     print(E)
        #     print("===="*30)

    
    with open(f"CE_BenchMark/Metadata/metadata.pkl", 'wb') as f:
        pickle.dump(metadata, f)


    return pipelines


now = datetime.datetime.now()
```
